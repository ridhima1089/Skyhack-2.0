```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load your datasets
# Replace 'call_data.csv' and 'reason.csv' with your actual dataset paths
calls_df = pd.read_csv('C:/UA/calls.csv')  # Contains call transcripts
reasons_df = pd.read_csv('C:/UA/reason.csv')  # Contains call reasons

# Display the first few rows of the reasons dataset
print("Reasons Data:")
print(reasons_df.head())

# Step 1: Analyzing Primary Call Reasons
# Check if 'primary_call_reason' exists in reasons_df
if 'primary_call_reason' in reasons_df.columns:
    # Create a frequency distribution of primary call reasons and select the top 12
    call_reason_counts = reasons_df['primary_call_reason'].value_counts().head(12)

    # Step 2: Visualization of Top 12 Primary Call Reasons
    plt.figure(figsize=(12, 6))
    call_reason_counts.plot(kind='bar', color='salmon')
    plt.title('Top 12 Primary Call Reasons')
    plt.xlabel('Primary Call Reasons')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Step 3: Recommendations for IVR Improvements
    print("Recommendations based on the top 12 most common call reasons:")
    for reason, count in call_reason_counts.items():
        print(f"- Enhance IVR options for: {reason} (occurrences: {count})")
else:
    print("Column 'primary_call_reason' not found in the reasons dataset.")
```

```python
def process_calls_for_reason(reason):
    # Filter for the specific call reason
    specific_calls = reasons_df[reasons_df['primary_call_reason'] == reason]
    specific_data = pd.merge(specific_calls, calls_df, on='call_id', how='inner')

    # Prepare the dataset
    transcripts = specific_data['call_transcript'].dropna()

    # Create a custom stop words list
    custom_stop_words = set(stopwords.words('english')) | set([
       'you', 'the', 'for', 'that', 'let', 'thank', 'call', 'agent', 'customer',
    'flight', 'would', 'look', 'get', 'fee', 'work', 'date', 
    'voucher', 'return', 'monday', 'friday', 'saturday', 'sir', 'agent', 'customer', 'day', 'help', 'flight', 'meeting', 'like', 'sir', 'delay', 'change',
    'earlier', 'time', 'bag', 'experi', 'return flight', 'okay','typing','could','sir','san francisco','ugh','took',
 'francisco','next','one','wednesday','united','tomorrow','today',
 'really','chicago','500','la','150',
 'sure',
 'customer',
 'chicago san',
 'francisco' ,

    'return date', 'thank', 'let'
    ])

    # Text Preprocessing
    def preprocess_text(text):
        # Tokenize and remove stop words
        words = text.split()
        filtered_words = [lemmatizer.lemmatize(word) for word in words if word.lower() not in custom_stop_words]
        return ' '.join(filtered_words)

    # Apply preprocessing to the transcripts
    cleaned_transcripts = transcripts.apply(preprocess_text)

    # Using bigrams and unigrams
    tfidf_bigram = TfidfVectorizer(ngram_range=(1, 2))  # Adjust ngram_range for unigrams and bigrams
    X_bigram = tfidf_bigram.fit_transform(cleaned_transcripts)

    # Fit KMeans clustering on bigrams
    num_clusters = 10  # Adjust the number of clusters here (e.g., 10)
    kmeans_bigram = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans_bigram.fit(X_bigram)

    # Print the cluster centroids for bigrams
    order_centroids_bigram = kmeans_bigram.cluster_centers_.argsort()[:, ::-1]
    terms_bigram = tfidf_bigram.get_feature_names_out()

    print(f"\nTop terms per cluster for '{reason}' using bigrams after filtering common words:")
    for i in range(num_clusters):
        print(f"\nCluster {i}:")
        for ind in order_centroids_bigram[i, :10]:  # Top 10 terms
            print(f' {terms_bigram[ind]}')

# List of reasons to process
reasons_to_process = ['IRROPS', 'Voluntary Change', 'Seating', 'Mileage Plus', 'Post-Flight', 'Communications',
    "Products and Services",
    "Upgrade",
    "Baggage",
    "Booking",
    "Checkout",
    "Check-In"
]

# Process each reason
for reason in reasons_to_process:
    process_calls_for_reason(reason)
```

```python
from sklearn.decomposition import PCA

def visualize_clusters_pca(X, kmeans_labels):
    # Step 1: Reduce dimensionality to 2D using PCA
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X.toarray())  # Convert sparse matrix to dense array

    # Step 2: Create a DataFrame for PCA results
    pca_df = pd.DataFrame(data=X_pca, columns=['PCA1', 'PCA2'])
    pca_df['Cluster'] = kmeans_labels

    # Step 3: Visualize the clusters
    plt.figure(figsize=(10, 6))
    for cluster in pca_df['Cluster'].unique():
        plt.scatter(pca_df[pca_df['Cluster'] == cluster]['PCA1'], 
                    pca_df[pca_df['Cluster'] == cluster]['PCA2'], 
                    label=f'Cluster {cluster}')

    plt.title('KMeans Clusters Visualization (PCA)')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

# Call the visualization function
visualize_clusters_pca(X_bigram, kmeans_bigram.labels_)
```
